# Creating a Column to show the distance of property from the city

```python

import numpy as np
 
def haversine_vectorize(lon1, lat1, lon2, lat2):
 
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
 
    newlon = lon2 - lon1
    newlat = lat2 - lat1
 
    haver_formula = np.sin(newlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(newlon/2.0)**2
 
    dist = 2 * np.arcsin(np.sqrt(haver_formula ))
    km = 6367 * dist #6367 for distance in KM for miles use 3958
    return km
```

```python
df['distance_from_seattle'] = haversine_vectorize(seattle_long,seattle_lat,df['long'],df['lat'])
```

```python
numerical_cols = []
for col in df.columns:
    if df[col].dtype in [np.int64, np.float64]:
        numerical_cols.append(col)

len(numerical_cols), numerical_cols[:5]
```

```python
check_col = ['price',
 'bedrooms',
 'bathrooms',
 'sqft_living',
 'sqft_lot','sqft_above',
 'sqft_basement',
  'Age_of_House',
 'Years_last_renovation',
 'distance_from_seattle',
 'distance_from_bellevue',
 'distance_from_a_city',
 'distance_from_airport']
```

```python
def remove_outliers(df, col, threshold = 3):
    zscores = scs.zscore(df[col])
    indices = np.abs(np.where(zscores > threshold))
    return indices[0]

check_col = numerical_cols

all_indices = []
for col in check_col:
    indices = remove_outliers(df, col, threshold = 3)
    all_indices.extend(indices)
all_indices = np.unique(all_indices)
```

```python
df.drop(all_indices,axis=0,inplace=True)
```
